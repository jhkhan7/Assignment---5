```python
import glob
import numpy 
import matplotlib
import pandas as pd
import matplotlib.pyplot 
import matplotlib.pyplot as plt
import numpy as np

dataset = sorted(glob.glob('data*.csv'))
print(dataset)
```

```python
def visualize(dataset_name):
    #read dataset  using pd.read_csv
    
    dataset1=pd.read_csv(dataset_name)
    #loop through all thye columns
    means=[]
    mins =[]
    maxs =[]
    features=[]
    dataset_width = dataset1.shape[1]
    for i in range(0, dataset_width):
        features.append(i)
    
    for columnname in dataset1:
        column_mean=dataset1[columnname].mean()
        column_min =dataset1[columnname].min()
        column_max =dataset1[columnname].max()
        
        means.append(column_mean)
        mins.append(column_min)
        maxs.append(column_max)
        
    plt.plot(features, means)
    plt.plot(features, mins)
    plt.plot(features, maxs)
    
    plt.legend(["avg", "mins", "maxs"])

    plt.show()
```

```python
def issue (dataset_name):
    df = pd.read_csv(dataset_name)
    if (df.iloc[:,0].max()==0 and df.iloc[:,20].max()==20):
        print ("I doubt the integrity of the datset")
    else:
        if sum(df.min(axis=0).to_list())==0:
            print("Why this is happening")
        else:
                print("Dataset looks good")
                
    
```

```python
for dataset_name in dataset[:4]:
    
    print(dataset_name)
    
    visualize(dataset_name)
    issue (dataset_name)
```
